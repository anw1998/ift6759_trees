```python
# Anni FasterCNN
ensemble=True
model_name = 'fastercnn'
if ensemble:
    run_name = 'RGB_baseline_ensemble_thres0.02'
    raw_exp_label_path = f'../fastercnn-pytorch-training-pipeline/outputs/ensemble/{run_name}'
else:
    run_name = 'RGB_baseline_thres0.5'
    raw_exp_label_path = f'../fastercnn-pytorch-training-pipeline/outputs/inference/{run_name}'
```

```python
chm_path = '../data/raw/RemoteSensing/CHM'
rgb_path = '../data/raw/RemoteSensing/RGB'
test_im_path = '../processed_data/test/images'
data_dir = f'./runs/clean_labels/{model_name}/{run_name}/submission'
os.makedirs(data_dir, exist_ok=True)
test_rgb = f'../processed_data/test/images'
raw_true_label_path = f'../processed_data/test/labels'

clean = f'./runs/clean_labels/{model_name}/{run_name}/'
out_dir = f'./scores/{model_name}/{run_name}'
os.makedirs(out_dir, exist_ok=True)
```

```python
# fastercnn
if ensemble:
    pred_path = sorted(glob(f'../fastercnn-pytorch-training-pipeline/outputs/ensemble/{run_name}/*.txt'))
else:
    pred_path = sorted(glob(f'../fastercnn-pytorch-training-pipeline/outputs/inference/{run_name}/*.txt'))
for pred in pred_path:
    f = os.path.basename(pred).split('.')[0]
    bbox_pred = np.loadtxt(pred)
    gdf = pd.DataFrame(bbox_pred, columns = ['minx', 'miny', 'maxx', 'maxy'])
    gdf['width'] = gdf['maxx'] - gdf['minx']
    gdf['length'] = gdf['maxy'] = gdf['miny']
    tmp = gdf[['minx', 'miny', 'width', 'length']].astype(int)
    tmp.to_csv(f'./runs/clean_labels/{model_name}/{run_name}/submission/{f}_submission.csv')
```

```python
gtf = pd.read_csv('../data/external/tmp_test.csv')
files = gtf['filename'].unique()
for f in files:
    tmp = gtf[gtf['filename'] == f]
    tmp['width'] = tmp['xmax'] - tmp['xmin']
    tmp['length'] = tmp['ymax'] - tmp['ymin']
    tmp2 = tmp[['xmin', 'ymin', 'width', 'length']]
    tmp2.columns = ['minx', 'miny', 'width', 'length']
    tmp2 = tmp2.reset_index(drop=True)
    tmp2.to_csv(f'./runs/clean_labels/{model_name}/{run_name}/submission/{f[:-4]}_ground.csv')
```
